```python
O=-snew.O + np.diag(snew.V)
```

```python
np.diag(O,snew.N+2)
```

```python
aband=np.zeros((snew.N,snew.N**2),float)
for i in range(snew.N):
    aband[i,i:]=np.diag(O,i)
```

```python
import scipy
En,Ev=scipy.linalg.eig_banded(aband)
```
